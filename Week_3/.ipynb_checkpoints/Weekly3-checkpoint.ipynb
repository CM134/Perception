```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

plt.gray()

img_left = cv2.imread("tsukuba_left.png")
img_left = cv2.cvtColor(img_left, cv2.COLOR_BGR2RGB)
img_gray_left = cv2.cvtColor(img_left, cv2.COLOR_RGB2GRAY)

plt.imshow(img_gray_left)

nose_left = cv2.imread("nose_left.png")
nose_left = cv2.cvtColor(nose_left, cv2.COLOR_BGR2RGB)
nose_gray_left = cv2.cvtColor(nose_left, cv2.COLOR_RGB2GRAY)

nose_right = cv2.imread("nose_right.png")
nose_right = cv2.cvtColor(nose_right, cv2.COLOR_BGR2RGB)
nose_gray_right = cv2.cvtColor(nose_right, cv2.COLOR_RGB2GRAY)

f, (ax_left, ax_right) = plt.subplots(1, 2, figsize=(12,12))
ax_left.imshow(nose_gray_left)
ax_right.imshow(nose_gray_right)
```

## Intermediate Step 1:

Compare the two equal-sized images and calc the sum of absolute difference. 

sad(x,y,d) = sum(I(x,y)-I(x,y-d)

```python
def SAD_slow(left,right):
    difference = 0
    for y in range(left.shape[0]):
        for x in range(left.shape[1]):
            difference = difference + np.abs(left[x,y].astype('float')-right[x,y].astype('float'))                               
    return difference

def SAD(left_img,right_img):
     return np.sum(np.abs(np.subtract(left_img, right_img, dtype=float)))


print(SAD(nose_gray_left,nose_gray_right))
#print(SAD_slow(nose_gray_left,nose_gray_right))
```

## Intermediate step 2:
finding the best match of nose_left in nose1 nose2 and nose3

```python
imagename = 'nose'
numbers = ['1.png','2.png','3.png']
dif = np.zeros(len(numbers))

for count, num in enumerate(numbers):
    nose_name = imagename+num
#     print(nose_name)
    nose = cv2.imread(nose_name)
    nose = cv2.cvtColor(nose, cv2.COLOR_BGR2RGB)
    nose_gray = cv2.cvtColor(nose, cv2.COLOR_RGB2GRAY)
    dif[count] = SAD(nose_gray_left,nose_gray)
    
print(dif)
idx = np.argmin(dif)
```

## Intermediate step 3:

We assume rectified image. Use nose_span.png and find the min difference ofer x.

```python
nose_span = cv2.imread("nose_span.png")
nose_span = cv2.cvtColor(nose_span, cv2.COLOR_BGR2RGB)
nose_gray_span = cv2.cvtColor(nose_span, cv2.COLOR_RGB2GRAY)

plt.imshow(nose_gray_span)
print(nose_gray_span.shape)
patch_width = nose_gray_left.shape[1]
print(patch_width)
```

```python
patch_width = nose_gray_left.shape[1]
dif = np.zeros(nose_gray_span.shape[1]-patch_width)

for disp in range(nose_gray_span.shape[1]-patch_width):
    # crop image in size of template
    # start from 0-7 end at end-7
    roi = nose_gray_span[:, disp:disp+patch_width]
    dif[disp] = SAD(nose_gray_left,roi)

    
disp_best = np.argmin(dif)
value_min = dif[disp_best]

print(disp_best,value_min)

# plot disparity graph
plt.plot(range(len(dif)),dif)
plt.xlabel('disparrity')
plt.ylabel('dissimilarity')
plt.show()


# plot the best images
f, (ax_left, ax_right) = plt.subplots(1, 2, figsize=(12,12))
ax_left.imshow(nose_gray_left)
ax_right.imshow(nose_gray_span[:,disp_best:disp_best+patch_width])
```

```python
def patch_matching(patch,span):
    patch_width = patch.shape[1]
    dif = np.zeros(span.shape[1]-patch_width)

    for disp in range(span.shape[1]-patch_width):
        # crop image in size of template
        # start from 0-7 end at end-7
        roi = nose_gray_span[:, disp:disp+patch_width]
        dif[disp] = SAD(nose_gray_left,roi)

    disp_best = np.argmin(dif)
    value_min = dif[disp_best]
    return disp_best

print(patch_matching(nose_gray_left,nose_gray_span))
```

```python
# import images

img_left = cv2.imread("tsukuba_left.png")
img_left = cv2.cvtColor(img_left, cv2.COLOR_BGR2RGB)
img_gray_left = cv2.cvtColor(img_left, cv2.COLOR_RGB2GRAY)


img_right = cv2.imread("tsukuba_right.png")
img_right = cv2.cvtColor(img_right, cv2.COLOR_BGR2RGB)
img_gray_right = cv2.cvtColor(img_right, cv2.COLOR_RGB2GRAY)

f, (ax_left, ax_right) = plt.subplots(1, 2, figsize=(12,12))
ax_left.imshow(img_gray_left)
ax_right.imshow(img_gray_right)
```

```python
patch_size = 7

left = img_gray_left
right = img_gray_right
disp_mat = np.zeros_like(left)

for y in range(left.shape[1]-patch_size):
    for x in range(left.shape[0]-patch_size):
            # get the patch
            patch = left[y:y+patch_size,x:x+patch_size]
            # get span
            span = right[y:y+patch_size,:]
           
            disp_mat[x,y] = patch_matching(patch,span)
            
print(disp_mat)
```
