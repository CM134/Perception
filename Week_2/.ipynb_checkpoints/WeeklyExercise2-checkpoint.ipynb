## Sparse Optical Flow

```python
# showing the features offsets for 1 sample back
cap = cv2.VideoCapture('Robots.mp4')
ret = 1
while ret:
    ret, frame = cap.read()
    if(ret):
        gray1 = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        ret, frame2 = cap.read()
        gray2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

        feat1 = cv2.goodFeaturesToTrack(gray1, maxCorners=100, qualityLevel=0.3, minDistance=7)
        feat2, status, error = cv2.calcOpticalFlowPyrLK(gray1, gray2, feat1, None)
        
        #print(feat1[1,:])
        #print(feat2[1,:])

        diff= np.abs((feat1-feat2).sum(axis=2))>1
        #print(diff)
        ft1 = feat1[diff]
        #print(ft1)
        ft2 = feat2[diff]
        for i in range(len(ft1)):
                cv2.line(frame2, (ft1[i][0], ft1[i][1]), (ft2[i][0], ft2[i][1]), (0, 0, 255), 2)
                cv2.circle(frame2, (ft2[i][0], ft2[i][1]), 5, (0, 255, 0), -1)
        cv2.imshow('image', frame2)
        if cv2.waitKey(20) & 0xFF == ord('q'):
            break


cap.release()

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np


# showing the offset for multiple samples back
cap = cv2.VideoCapture('Robots.mp4')
ret = 1
gray = np.zeros((1,1080,1920),dtype=np.uint8)
n=0
sz =3
trace_ft1 = np.zeros((1,2))
trace_ft2 = np.zeros((1,2))

while ret:
    ret, frame = cap.read()
    frame_g = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    gray = np.append(gray,np.expand_dims(frame_g, 0),axis=0)
    if(ret and n>=sz):
        gray = gray[1:]
        feat1 = cv2.goodFeaturesToTrack(gray[0], maxCorners=100, qualityLevel=0.3, minDistance=7)
        feat2, status, error = cv2.calcOpticalFlowPyrLK(gray[0], gray[-1], feat1, None)

        diff= np.abs((feat1-feat2).sum(axis=2))>1
        ft1 = feat1[diff]
        ft2 = feat2[diff]
        # print(ft1.shape)

        # just for fun, saving the traces of where it was for the whole video
        trace_ft1 = np.concatenate((trace_ft1,ft1))
        trace_ft2 = np.concatenate((trace_ft2,ft2))
        #print(trace_ft1.shape)


        frame_plot = np.copy(frame)
        for i in range(len(ft1)):
            cv2.line(frame_plot, (ft1[i][0], ft1[i][1]), (ft2[i][0], ft2[i][1]), (0, 255, 0), 2)
            cv2.circle(frame_plot, (ft2[i][0], ft2[i][1]), 5, (0, 255, 0), -1)

        cv2.imshow('image', frame_plot)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        cv2.waitKey(1)

    n = n+1

cap.release()

cv2.destroyAllWindows()
```

```python
# some lefover code from some tutorial

def get_direction(self, frame1, frame2, show=False):
        frame1 = cv2.resize(frame1, (self.width, self.height))
        frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
        frame2 = cv2.resize(frame2, (self.width, self.height))
        frame2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

        flow = cv2.calcOpticalFlowFarneback(frame1[self.height_start:self.height_end],
                                            frame2[self.height_start:self.height_end], None, 0.5, 3, 15, 1, 5, 1.2, 0)
        flow_avg = np.median(flow, axis=(0, 1))  # [x, y]

        move_x = -1 * flow_avg[0]
        move_y = -1 * flow_avg[1]

        if show:
            hsv = np.zeros((self.height_end - self.height_start, self.width, 3))
            hsv[...,1] = 255
            mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])
            hsv[..., 0] = ang * 180 / np.pi / 2
            hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
            bgr = cv2.cvtColor(np.array(hsv).astype(np.uint8), cv2.COLOR_HSV2BGR)

            cv2.imshow('opt_flow', bgr)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                print('User Interrupted')
                exit(1)

        return move_x, move_y 
```
